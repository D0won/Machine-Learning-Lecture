### 📝 문제 9: 분산과 표준편차 마스터하기! (난이도: ⭐⭐⭐)

**📊 개념 정리:**
- **분산(variance)**: 데이터가 평균에서 얼마나 흩어져 있는지를 나타내는 값
- **표준편차(standard deviation)**: 분산의 제곱근으로, 분산보다 해석하기 쉬운 값
- 분산과 표준편차가 클수록 → 데이터가 평균에서 많이 흩어져 있음
- 분산과 표준편차가 작을수록 → 데이터가 평균 주변에 모여 있음

1. **스칼라 + 배열**: 어떤 크기든 가능 (예: 배열 + 5)
2. **같은 차원**: 행/열 개수가 일치 (예: (3,4) + (3,4))  
3. **차원 확장**: 한쪽이 1인 차원 (예: (3,1) + (1,4) → (3,4))
4. **벡터 확장**: 1차원이 2차원으로 (예: (3,4) + (4,))

**🏪 상황:**
편의점 체인점에서 상품별, 지점별 가격 정책을 관리합니다.
