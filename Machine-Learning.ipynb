# 머신러닝이란?

## 전체적인 개요

- 먼저, 인공지능이란 우리 인간이 가진  판단, 사고, 예측 등의 지적 능력을 기계를 통해 구현하는 기술을 뜻한다.

- 머신러닝은 인공지능의 한 분야로 기계가 스스로 학습하여 새로운 정보를 얻어내거나 예측하는 기술이다. 요즘에 각광받는 기술로
많은 인공지능 개발자들이 이 분야를 연구하고 있다.

- 딥러닝은 머신러닝의 한 분야로 우리 인간이 가진 뇌 세포인 뉴런을 모방한 인공 신경망을 이용하여 머신러닝을 구현하는 기술이다. 이는 나중에 나오는 내용으로 추후에 설명하곘다.

## 머신러닝 뜻

- 머신러닝이란 기계가 수많은 데이터로 학습하여 스스로 데이터 간의 규칙을 발견하고, 이 규칙을 새로운 데이터에 적용하여 정보를 얻어내거나 예측하는 기술이다.

- 예를 들어, 고양이와 강아지를 분류하는 인공지능을 만든다고 하자. 이때 수많은 종류의 고양이 사진 10,000장과 수많은 종류의 강아지 사진 10,000장을 인공지능 모델에 학습시킨다. 학습이 완료된 인공지능 모델이 길고양이 사진을 보고 이를 고양이라고 판단하게끔 만드는 것이 머신러닝이다.

## 머신러닝 원리

- 머신러닝의 원리는 데이터의 특성과 패턴을 학습하여 모델을 생성한 후에 새로운 입력 데이터의 결과를 예측시키는 것이다.


- 이미지를 통해 살펴보면 입력 변수에 입력 데이터를, 출력 변수에 출력 데이터를 집어넣어서 기계가 직접 데이터의 특성과 원리를 학습하여 수학적 모델을 만드는 것이다.



- 위 표와 같이 기계가 x(입력 데이터), y(출력 데이터) 데이터의 특성과 패턴을 학습하여 y = ㅁ * x + o 라는 모델에서 ㅁ와 o를 찾는 것이다.

- 위 표를 보고 우리가 ㅁ와 o을 찾아보면 ㅁ는 2, o는 1임을 알 수 있다. 이렇듯 기계도 마찬가지로 이러한 과정을 거쳐서 ㅁ와 o을 찾아 y = 2x + 1이라는 수학적 모델을 만들고, 새로운 x값에 대한 결과를 예측한다.

## 참고) 인간과 기계의 학습 차이

- 인간은 경험을 토대로 미래를 예측하고 기게는 데이터 학습을 통해 결과나 미래를 예측한다.
- 따라서 경험을 토대로 예측하는 것보다 정확한 데이터를 학습하여 예측하는 것이 더 정확하므로 인간의 미래 예측보다 기계의 미래 예측이 더욱 뛰어나다.

## 정통적인 프로그래밍과 머신러닝의 차이

- 전통적인 프로그래밍이란? 지금까지 우리가 파이썬으로 프로그래밍(코딩)을 하던 것을 정통적인 프로그래밍이라고 한다.

- 우리가 지금까지 했던 프로그래밍은 데이터와 규칙을 정해 주면 새로운 데이터를 규칙에 적용하여 결과를 출력시킨다.

- 반면에 머신러닝은 주어진 데이터를 학습하면서 데이터 간의 관계를 가장 잘 설명하는 규칙을 스스로 발견한다.

- 정통적인 프로그래밍의 경우 우리가 직접 데이터와 규칙을 정하고 이에 따라 프로그래밍을 해서 새로운 데이터에 대한 결과를 출력시키고

- 머신러닝 같은 경우에는 데이터와 출력만을 기계에 학습시켜 기계가 스스로 규칙을 찾는 것이다.

- 예시를 들어보자면 우리가 만약에 제곱수를 구하는 프로그램을 짜본다고 하자. 그렇다면 우리가 제곱수를 찾기 위해서는 x^2 이라는 규칙을 적용하여 프로그래밍을 하면 된다. 따라서 전통적인 프로그래밍은 3이라는 입력 데이터와 x^2 이라는 규칙을 주어질 때, 그 결과로 9를 출력한다.

- 반면에 머신러닝은 입력 데이터와 출력만을 기계에 입력하면 기계가 알아서 규칙을 찾는다.

- 정리해보자면, 우리가 지금까지 헀던 프로그래밍은 목표가 정확한 출력값을 구하는 것이고, 머신러닝의 목표는 주어진 데이터와 출력에 대한 규칙을 구하는 것이다.

- 따라서 머신러닝 같은 경우에는 방대한 양의 데이터 예를 들어 1억 개 - 10억 개 정도의 데이터들의 규칙을 찾는 데 유용하다.
(인간도 찾으라면 찾을 수 있겠지만, 아마 죽을 때까지 못 찾을 수 있다. 반면에 기계에 이들의 데이터를 학습 시키는 경우에 규칙을 매우 빠르게 찾을 수 있다. 데이터 수에 따르겠지만 아마 3일 내로 규칙을 찾을 것이다.)

## 머신러닝 학습 방법 3가지

- 머신러닝에는 3가지의 학습 방법이 존재한다.
- 지도학습, 비지도 학습, 그리고 강화학습이 머신러닝 학습 방법의 3가지 방법이다.

### 지도학습

- 지도학습의 경우 레이블(정답 또는 출력 데이터)과 입력 데이터를 함계 제공하는 학습 방법이다.



- 위 그림을 참고하면 장미 사진과 해바라기 사진을 보여주며 선생님이 이 사진은 장미고 이 사진은 해바라기야 라고 가르쳐주는 모습을 볼 수 있다. 이 때 입력 데이터는 장미 사진 그리고 해바라기 사진이고 선생님이 이것의 이름을 가르쳐주는 것이 바로 레이블 즉 정답 또는 출력 데이터를 가르쳐 주는 것이다.

- 따라서 입력 데이터와 그에 따른 정답을 가르쳐주는 것이 지도학습이다.

### 비지도학습

- 비지도학습은 앞서 지도학습과 달리 레이블이 없는 데이터를 분석하여 데이터의 특성과 관계를 찾게 하는 학습 방법이다.
오른 쪽 그림과 같이 장미 꽃 그림과 해바라기 꽃 그림을 기계한테 주면 기계가 알아서 데이터의 특성에 따라 분류해서 그룹을 만든다. 이와 같은 경우에는 빨간 특성 및 모양을 보고 장미 꽃 그룹을 노란색 특성 및 모양을 보고 해바라기 그룹을 지을 수 있다.

- '비'지도학습이므로 이 때는 입력 데이터만 주어질 뿐 정답은 주어지지 않는다.

### 강화학습

- 강화학습이란 여러번의 시뮬레이션을 통해 어떤 행동에 대한 보상이 최대가 되도록 강화시키는 학습 방법이다.
- 예를 들어, 기계에게 벽돌 게임을 시킨다고 하자 1개의 벽돌을 부섰다면 보상을 주지 않거나 매우 조금 주고 4개 이상의 벽돌을 부셨다면 보상을 최대로 줘 이 게임을 순조롭게 마스터하게끔 만드는 것이다.
- 이는 지도학습과 비지도학습과는 완전히 다른 방향이기 때문에 이번에는 다루지 않을 예정이다.

## 학습 방식에 따른 머신러닝 구분

### 분류

- 분류는 지도학습의 방식 중 하나로 새로 입력된 데이터가 어디에 속하는지를 분류하는 것이다.
- 따라서 출력값이 범주로 나뉘어지는 값(이산형 데이터)인데 이는 여러 그룹이 존재할 때 그 그룹등 중 하나를 고르는 것을 뜻한다.

- 예를 들어, 이메일이 스팸인지 아닌지를 분류하거나 시험 결과를 보고 합격인지 불합격인지를 판단하거나 앞서 예시를 든 것 처럼 사진을 보고 강아지인지 고양이인지 알아내는 것이다.

- 과정을 보면 정답이 있는 데이터 셋을 기계에다 학습을 시켜 분류 모델을 만들고 새로운 데이터에 대해서 어디에 속할지 확률을 구한 후 확률이 높은 것으로 분류시킨다.

### 회귀

- 회귀는 지도학습의 한 방식으로, 새로운 입력 데이터에 대해 연속적인 숫자 값을 예측하는 것이다.
- 출력값은 연속형 데이터로 나타나며, 이는 그래프 상의 특정 점으로 표현될 수 있다.

- 예를 들어, 집 크기를 보고 집값이 얼마일지 예측하는 것, 또는 공부 시간으로 시험 점수를 예측하는 것, 또한 키와 몸무게를 보고 체지방률 추정하는 것이다.

- 과정을 보면 정답이 있는 데이터 셋을 학습 시켜서 예측 모델을 만들고 새로운 데이터에 대해서 예측 결과를 내놓는 것이다.

### 군집화

- 군집화란 데이터의 특성과 패턴 등을 파악한 후에 그룹을 짓는 거야(군집화)

- 레이블(정답)이 없는 데이터 셋을 기계에 입력하여, 기계가 데이터의 특성과 패턴을 분석해 각 데이터를 그룹으로 나누는 것

- 예를 들어 먼저, 지도학습은 오리 사진들과 토끼 사진, 고슴도치 사진을 주고 이에 대해서 오리 사진은 오리, 토끼 사진은 오리 아님. 이라고 정답을 지어주고 이를 토대로 기계에 학습을 시키는 것. 따라서 새로운 오리 이미지를 분류 모델에 주었을 때 분류 모델이 '오리' 라고 판단하는 것이다.

- 반면에 비지도학습은 오리 사진들과 토끼 사진, 고슴도치 사진만 있고 이에 대한 정답이 없는 채로 기계에다 학습을 시키는 것이다. 그러면 기계가 데이터의 특성과 패턴을 스스로 찾아서 오리는 오리끼리 그룹짓고 나머지는 나머지대로 그룹을 짓게 만든다.

## 지도,비지도,강화 학습 정리

- 지도학습은 레이블이 있는 즉, 정답이 있는 데이터를 학습시켜서 데이터의 결괏값이나 미래를 예측하는 것
- 비지도학습은 레이블이 없는 즉, 정답이 없는 데이터를 기계에다 던져주고 기계가 직접 데이터의 숨겨진 패턴 및 관계를 분석하여 그룹을 짓게 하는 것
- 강화학습은 데이터가 필요없고 기계의 행동에 따라 보상을 주는 것

## 지도 학습 과정

- 지도학습 과정은 크게 학습 단계와 예측 단계로 구분한다.
- 학습 과정은 인공지능 모델 생성 과정을 뜻하고, 예측 단계는 만들어진 모델로 새로운 데이터에 대한 예측을 수행하는 것이다.

#### 학습 단계
1. 먼저 데이터를 수집한다.
2. 수집한 데이터를 기계가 학습할 수 있도록 전처리(데이터의 품질을 높이고 기계가 이해할 수 있도록 데이터를 정리하거나 변환하는 작업을 한다. 예를 들어, 누락된 데이터를 채우거나, 필요 없는 데이터를 제거하거나, 데이터를 숫자로 변환하는 과정을 포함한다.)한 후, 입력 데이터와 정답이 있는 훈련 데이터 셋과 테스트 데이터 셋으로 나눈다.
3. 훈련 데이터 셋을 이용하여 모델을 학습시킨다.
4. 테스트 데이터 셋을 이용해서 예측한 값과 정답을 비교한다.

#### 예측 단계
1. 새로운 데이터를 기계에 맞게 전처리를 한다.
2. 전처리한 데이터를 학습된 모델에 입력을 한다.
3. 결과값을 예측한다.
