### 8. 결손 데이터를 처리하기 위해 다음 문제들을 푸시오.

#### 8-2. 혈압, 콜레스테롤, 최대 심박수 칼럼들에 있는 null 값들을 다 평균값으로 번경하시오.

### 9. 람다식을 이용하여 혈압을 다음 표를 기준으로 나눠 새로운 칼럼을 생성하시오.

| 분류 | 수축기 혈압 기준 |
| :--- | :--- |
| **저혈압** | $90 \text{ mmHg}$ 미만 |
| **정상 혈압** | $90 \text{ mmHg}$ ~ $119 \text{ mmHg}$ |
| **주의 혈압** | $120 \text{ mmHg}$ ~ $129 \text{ mmHg}$ |
| **고혈압 1기** | $130 \text{ mmHg}$ ~ $139 \text{ mmHg}$ |
| **고혈압 2기** | $140 \text{ mmHg}$ 이상 |

```python
def classify_blood_pressure(bp):
    if bp < 90:
        return '저혈압'
    elif 90 <= bp <= 119:
        return '정상 혈압'
    elif 120 <= bp <= 129:
        return '주의 혈압'
    elif 130 <= bp <= 139:
        return '고혈압 1기'
    else :
        return '고혈압 2기'

heartDiseaseData['혈압 분류'] = heartDiseaseData['혈압'].apply(classify_blood_pressure)
heartDiseaseData.head()
```
