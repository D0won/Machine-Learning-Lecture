## groupby() 함수

`groupby()`는 데이터를 특정 기준으로 그룹화하고, 각 그룹에 대한 통계량(평균, 합계 등)을 계산하는 데 사용하는 판다스의 함수

---

`groupby()` 함수는 "분할(split)", "적용(apply)", "결합(combine)"의 세 단계를 거쳐 작동

1.  분할(Split): 특정 열(예: 'Embarked' 항구, 'Pclass' 객실 등급)의 고유값들을 기준으로 데이터를 여러 개의 작은 그룹으로 나눔
2.  적용(Apply): 각 그룹에 대해 원하는 연산(예: 평균 `mean()`, 합계 `sum()`, 최댓값 `max()`, 개수 `count()` 등)을 적용
3. 결합(Combine): 각 그룹에 적용된 연산의 결과를 하나의 새로운 데이터프레임으로 결합하여 반환

예를 들어, 타이타닉 데이터에서 `groupby('Pclass')`를 사용하면 1등석, 2등석, 3등석 승객 그룹으로 나눌 수 있다. 여기에 `.mean()`을 적용하면 각 객실 등급별 승객의 평균 나이, 평균 운임 등을 쉽게 구할 수 있다.
